```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import cross_val_score
```

# Problem 6 Home Work 2

```python
Train_X = pd.read_csv('Problem 6/data/trainx.csv',header=None)
Train_y = pd.read_csv('Problem 6/data/trainy.csv',header=None)
Test_x = pd.read_csv('Problem 6/data/testx.csv',header=None)
Test_y = pd.read_csv('Problem 6/data/testy.csv',header=None)
```

```python
Train_y = sklearn.utils.column_or_1d(Train_y, warn=True)
Test_y = sklearn.utils.column_or_1d(Test_y, warn=True)
```

Creating the model and fit the data into it

```python
model = LogisticRegression(solver='liblinear', random_state=0)
model.fit(Train_X, Train_y)
```

```python
predictions = model.predict(Test_x)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(Test_y,predictions))
print(classification_report(Test_y,predictions))
```
