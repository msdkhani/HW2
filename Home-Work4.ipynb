# Problem 6

```python
TrainX = pd.read_csv('Data for Problem 6/trainx.csv',header=None)
TrainY = pd.read_csv('Data for Problem 6/trainy.csv',header=None)
Testx = pd.read_csv('Data for Problem 6/testx.csv',header=None)
Testy = pd.read_csv('Data for Problem 6/testy.csv',header=None)
```

```python
plt.plot(TrainX,TrainY,'o')
```

```python
p6_model = LinearRegression().fit(TrainX,TrainY)
# we get our coefficents as W.T
wT = np.array(p6_model.coef_[0][0])
# intercept
intercept = p6_model.intercept_
```

```python
p6_predictions = p6_model.predict(Testx)
print(mean_squared_error(Testy,p6_predictions))
```

```python
plt.scatter(Testx, Testy,  color='black')
plt.plot(Testx, p6_predictions, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
# Linear regression Function
def fx(w,x,intercept):
    #y is our output
    y = np.dot(w,x)+intercept
    return np.array(y)
```

```python
xTrain=TrainX.values
yTrainPrediction = fx(wT,xTrain,intercept)
print('Mean squared error in Train Data is :',mean_squared_error(TrainY , yTrainPrediction))
```

```python
z = np.dot(wT,xTrain)
print('Mean squared error in Train Data without using intercept is :',mean_squared_error(z , yTrainPrediction))
```

```python
plt.scatter(TrainX, TrainY,  color='black')
plt.plot(TrainX, yTrainPrediction, color='blue', linewidth=4,label='Using Intercepet')
plt.plot(TrainX, z, color='red', linewidth=4,label='without Intercepet')
plt.xticks(())
plt.yticks(())

plt.legend()
plt.show()
```

```python
xTest=Testx.values
yTestPrediction = fx(wT,xTest)
print('Mean squared error in Test Data is :',mean_squared_error(Testy , yTestPrediction))
```

```python
plt.scatter(Testx, Testy,  color='black')
plt.plot(Testx, yTestPrediction, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

### In Problem 6 the data have the linear shape, we can easily estimate the value of test with the reasonable MSE, as we can see our function works well in both train and test data

# Problem 7

```python
trainX = pd.read_csv('Data for Problem 7/trainx.csv',header=None)
trainY = pd.read_csv('Data for Problem 7/trainy.csv',header=None)
testx = pd.read_csv('Data for Problem 7/testx.csv',header=None)
testy = pd.read_csv('Data for Problem 7/testy.csv',header=None)
```

```python
plt.plot(trainX,trainY,'o')
```

```python
p7_model = LinearRegression().fit(TrainX,TrainY)
WT = np.array(p7_model.coef_[0][0])
p7Intercept = p7_model.intercept_
```

```python
p7_predictions = p7_model.predict(testx)
print(mean_squared_error(testy,p7_predictions))
```

```python
plt.scatter(testx, testy,  color='black')
plt.plot(testx, p7_predictions, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
XTrainValue=trainX.values
YTrainValue = fx(WT,XTrainValue,p7Intercept)
```

```python
print(mean_squared_error(trainY , YTrainValue))
```

```python
plt.scatter(trainX, trainY,  color='black')
plt.plot(trainX, YTrainValue, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

```python
XTestValue=testx.values
YTestValue = fx(WT,XTestValue,p7Intercept)
```

```python
print(mean_squared_error(testy , YTestValue))
```

```python
plt.scatter(testx, testy,  color='black')
plt.plot(testx, YTestValue, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

### In problem 7 we cannot get the good estimation based on linear model, because our data is not linear, hence the model do not get the good MSE in this problem
