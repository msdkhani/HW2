# Import Libraries

```python
import numpy as np
import pandas as pd
```

#### Creating the graph

```python
g = { 
        1: [2],
        2: [3],
        3: [1,4,5],
        4: [2,6],
        5: [],
        6: [7],
        7: [8],
        8: [6]
}
```

```python
N = len(g)
print("graph length is :",N)
```

```python
A = np.zeros((len(g),len(g)))
for i in range(1,9):
    for j in range(1,9):
        if len(g[i]) != 0 and j in g[i]:
            A[j-1,i-1] = 1/len(g[i])
pd.DataFrame(A)
```

#### we set our beta to 0.8 and for our threshold we use epsilon 0.01. In this project, our specific pages are 3 and 4.


```python
beta = 0.8
epsilon = 0.01
# number of nodes
N = 8 
# Specific pages
s = [3,4]
s_length = len(s)
```

# First we code Standard PageRank

```python
B = np.zeros((8,8))
B[:] = 1/N
B
```

```python
Al = beta * A + ((1-beta) * B)
pd.DataFrame(Al)
```

```python
r = np.zeros((1,8)).T
r[:] = 1/N
r
```

```python
diff = np.infty
iterration = 0
while diff > epsilon:
    r_old = r
    new_r = r = np.dot(Al,r)
    diff = sum(abs(new_r - r_old))
    iterration +=1
print("Done", iterration)
r
```

# Page Specific

#### for page specific we just need to calculate the Matrix "Al" For specific pages number 3 and 4


```python
AS = np.zeros((8,8))
for i in range(N):
    for j in range(N):
        # we set i+1 because the index is starting from 0
        if i+1 in s:
            AS[i,j] = beta * A[i,j] + ((1-beta)/s_length)
        else:
            AS[i,j] = beta * A[i,j]
pd.DataFrame(AS)
```

```python
R = np.zeros((1,8)).T
R[:] = 1/N
```

```python
different = np.infty
iteration = 0
while different > epsilon:
    R_old = R 
    new_R = R = np.dot(AS,R)
    different = sum(abs(new_R - R_old))
    iteration+=1
print("Completed after",iteration, "iteration")
R
```
