```python
import pandas as pd
import numpy as np

import keras
from tensorflow.keras import models,layers

import sklearn
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import OneHotEncoder
```

# Problem 2

```python
Train_X = pd.read_csv('Data for Problem 2/data/trainx.csv',header=None)
Train_y = pd.read_csv('Data for Problem 2/data/trainy.csv',header=None)
Test_x = pd.read_csv('Data for Problem 2/data/testx.csv',header=None)
Test_y = pd.read_csv('Data for Problem 2/data/testy.csv',header=None)
```

```python
problem2_model = MLPClassifier(random_state=1, max_iter=300,hidden_layer_sizes=100)
problem2_model.fit(Train_X,Train_y)
```

```python
p2_predictions = problem2_model.predict(Test_x)
print(confusion_matrix(Test_y,p2_predictions))
print(classification_report(Test_y,p2_predictions))
```

#### in this problem, we do not get better accuracy than Logistic Regression(for MLP model). Basically using Neural Network is not the best way always. In some data using some traditional machine learning model is more sufficient.

```python
problem2_model.coefs_[0]
```

# problem 3

```python
train_X = pd.read_csv('Data for Problem 3/data/trainx.csv',header=None)
train_y = pd.read_csv('Data for Problem 3/data/trainy.csv',header=None)
test_x = pd.read_csv('Data for Problem 3/data/testx.csv',header=None)
test_y = pd.read_csv('Data for Problem 3/data/testy.csv',header=None)
```

```python
problem3_model = MLPClassifier(random_state=1, max_iter=300,hidden_layer_sizes=100)
problem3_model.fit(train_X,train_y)
```

```python
p3_predictions = problem3_model.predict(test_x)
print(confusion_matrix(test_y,p3_predictions))
print(classification_report(test_y,p3_predictions))
```

# Problem 4

```python
TrainX = pd.read_csv('Data for Problem 4/train_input.csv',header=None)
TrainY = pd.read_csv('Data for Problem 4/train_labels.csv',header=None)
TestX = pd.read_csv('Data for Problem 4/test_input.csv',header=None)
TestY = pd.read_csv('Data for Problem 4/testy.csv')
```

```python
problem4_model = MLPClassifier(random_state=2, max_iter=500,hidden_layer_sizes=200,epsilon=0.00003,solver='adam')
problem4_model.fit(TrainX,TrainY)
```

```python
p4_predictions = problem4_model.predict(TestX)
print(confusion_matrix(TestY,p4_predictions))
print(classification_report(TestY,p4_predictions))
```
